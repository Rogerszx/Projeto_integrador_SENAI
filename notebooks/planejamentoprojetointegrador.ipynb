# 1. APRESENTAÇÃO DO PROJETO

Este documento propõe a criação de um sistema de controle de acesso na empresa via reconhecimento facial, em substituição ao modelo de crachás físicos e/ou senhas. A importância dessa substituição se deve,  principalmente, à vulnerabilidade dos modelos mais tradicionais, que podem facilmente ser burlados por meios como espionagem, roubo ou compartilhamento indevido, além da praticidade e modernização da biometria facial.  

# 2. DEFINIÇÃO DO PROBLEMA

O modelo atual de segurança física permite que as regras de acesso sejam burladas, uma vez que objetos ou senhas podem ser compartilhados, esquecidos ou extraviados, causando brechas graves na segurança. A implementação do reconhecimento facial resolverá esse problema ao transformar a própria face do colaborador em sua credencial intransferível. Isso beneficiará a empresa, que garantirá a autenticidade dos acessos, e os colaboradores, que ganharão agilidade na verificação e não precisarão mais se preocupar em portar itens físicos.

# 3. OBJETIVO DO PROJETO

O objetivo é: Um painel fará a leitura facial do rosto do colaborador, onde através da imagem escaneada estabelecerá uma comparação com o banco de dados específico para o setor que está tentando ser acessado. Após a leitura, no painel será retornado o valor "ACESSO AUTORIZADO" ou "ACESSO NEGADO" dependendo do grau de permissão encontrada no banco de dados.

# 4. TIPO DE PROBLEMA DE IA

O projeto se classifica como um problema de Visão Computacional, sendo uma tarefa de classificação binária, onde apenas dois resultados podem ser retornados(Acesso autorizado ou negado).
A solução proposta adotará a abordagem de verificação facial 1:1, onde o processo funcionará como um tipo de autenticação multifatorial(MFA): o usuário previamente informará sua identificação, como seu CPF ou número de matrícula, sendo realizada inicialmente uma verificação lógica rápida para aquela credencial e, caso ela sequer possua permissão para o setor, o acesso é negado imediatamente. Caso a credencial digitada seja válida, a IA atuará comparando a imagem capturada com o template associado à identificação digitada.

# 7. MÉTRICAS DE AVALIAÇÃO

Serão utilizadas métricas específicas para o controle de acesso, como:
- A taxa de falsos positivos, que medirão a frequência com que o sistema pode autorizar indevidamente o acesso de uma pessoa não correspondente à credencial informada, sendo necessária que essa taxa esteja o mais próxima possível de zero.

# 8. DESAFIOS E LIMITAÇÕES

Os principais desafios técnicos para a implementação dessa solução são variáveis que muitas vezes não podem ser controladas, como variação de iluminação do ambiente e obstruções faciais, como uso de óculos, máscaras ou chapéus/bonés. Em questão de segurança, um possível invasor poderia utilizar uma foto, seja ela impressa ou em algum aparelho eletrônico para burlar a verificação, sendo necessária uma futura implementação de um algoritmo de prova de vida, conhecido como liveness detection(ex: a necessidade de piscar os olhos uma determinada quantidade de vezes para que a verificação seja completa).
